## Theoretic Bounds

Set up the functions as described in the paper

```python
i,j,k = var('i,j,k') # throwaway variables

# standard functions
FR(r,l) = 2*r*(2*r-1)**(l-1)
CR(r,l) = (2*r-1)**l + 1 + (r-1)*(1+(-1)**l)

CR2(r,l1,l2) = sum(CR(r,i), i,l1,l2)

# Lemma 3.1: upper bound on NC1
nc1TightBound(lam,r,l,m) = \
    2*l*(l - 2*ceil(lam*l) - 2) * FR(r,ceil(lam*l)) * (2*r-1)**(l-2*ceil(lam*l)) \
    + \
    sum( \
      l * CR(r,k) * (2*r-1)**(l-ceil(lam*l) - k) \
    , k,1,ceil(lam*l))

# Lemma 3.2: upper bound on NC2
nc2TightBound(lam,r,l1,l2,m) = sum(sum( \
      2*i*j \
      * FR(r,ceil(lam*min_symbolic(i,j))) \
      * (2*r-1)**(i+j - 2*ceil(lam*min_symbolic(i,j))) \
  ,j,l1,l2) ,i,l1,l2)

# misc: lower bound on probability
lowerBound(lam,r,l1,l2,m) = 1 \
    - m * sum(nc1TightBound(lam,r,i,m), i,l1,l2) / CR2(r,l1,l2) \
    - m * (m-1) * nc2TightBound(lam,r,l1,l2,m) / (CR2(r,l1,l2)**2)

# Lemma 3.3: test if has zero probabilty
#  NOTE: this can still return false when the probability is zero
knownZeroProp(lam,r,l,m) = ( FR(r,ceil(lam*l)) - 2*m*l < 0 )
knownZeroProp2(lam,r,l,m) = ( CR(r,l) - 4*m*l*(r-1)*(2*r-1)**(l-ceil(lam*l)-1) < 0 )

# Lemma 3.3: upper bound on probability
upperBound(lam,r,l,m) = product( min_symbolic(\
    CR(r,l) - 4*i*l*(r-1)*(2*r-1)**(l-ceil(lam*l)-1), \
    (FR(r,ceil(lam*l)) - 2*i*l) * (2*r-1)**(l-floor(1/lam)*ceil(lam*l)) * \
    product( min_symbolic( (2*r-1)**ceil(lam*l), \
        FR(r,ceil(lam*l)) \
          - 2*i*l \
          - 2*((k-2)*ceil(lam*l) + 1 + l - floor(1/lam)*ceil(lam*l)) \
    ) ,k,2,floor(1/lam)) \
) ,i,0,m-1) / ( CR(r,l)**m )

# Lemma 4.1: check if small cancelation is known to hold
knownSmallCancelation(p,lam,r,l,m) = ( \
  (1-p) * (2*r-1)**(1+lam*l) - 8 * r * m**2 * l**2 >= 0 )
# also, a rough estimate of the probability
lowerBoundRough(lam,r,l,m) = \
  ( (2*r-1)**(1+lam*l) - 8 * r * m**2 * l**2 ) /  ( (2*r-1)**(1+lam*l) )
```

## Cached Versions

```python

@func_persist
def upperBoundCached(lam,r,length,m,dig = 8):
    return upperBound(lam,r,length,m).full_simplify().n(digits=dig)


@func_persist
def lowerBoundCached(lam,r,l1,l2,m,dig = 8):
    return max_symbolic(0,lowerBound(lam,r,l1,l2,m)).full_simplify().n(digits=dig)


def upperBoundCached2(lam,r,length,m,dig = 8):
    if (bool(knownZeroProp(1/6,r,length,m))) or (bool(knownZeroProp2(1/6,r,length,m))):
        return  0
    return upperBoundCached(lam,r,length,m,dig)

def lowerBoundCached2(lam,r,l1,l2,m,dig = 8):
    return lowerBoundCached(lam,r,l1,l2,m,dig)
```

## Experimental Data

Read the data from the files `../data/[date-hour]/data_[r]_[m]_[l1]_[l2].txt` into ``data_set``

```python
import os
import subprocess

data_set = {}

for at_time in os.listdir('../data/'):

    for meta in os.listdir('../data/'+at_time):
        file_name = os.path.join('../data/'+at_time,meta)

        line = subprocess.check_output(['tail', '-1', file_name])

        new_data = [int(v) for v in line.split(",")[:int(2)]]

        if meta in data_set:
            old_data = data_set[meta]
            data_set[meta] = [sum(x) for x in zip(old_data, new_data)]
        else:
            data_set[meta] = new_data
```

## Full Data Set

```python
import os
import re
import subprocess

full_data_set = {}

def get_data(r,m,l1,l2):
    data = full_data_set
    
    if int(r) in data:
        data = full_data_set[int(r)]
    else:
        return None
    
    if int(m) in data:
        data = full_data_set[int(m)]
    else:
        return None
    
    if int(l1) in data:
        data = full_data_set[int(l1)]
    else:
        return None
    
    if int(l2) in data:
        data = full_data_set[int(l2)]
    else:
        return None
    
    return data

def add_data(r,m,l1,l2, datum):
    
    if int(r) not in full_data_set:
        full_data_set[int(r)] = {}
    
    if int(m) not in full_data_set[int(r)]:
        full_data_set[int(r)][int(m)] = {}
    
    if int(l1) not in full_data_set[int(r)][int(m)]:
        full_data_set[int(r)][int(m)][int(l1)] = {}
    
    if int(l2) not in full_data_set[int(r)][int(m)][int(l1)]:
        full_data_set[int(r)][int(m)][int(l1)][int(l2)] = datum
        return
    
    old = full_data_set[int(r)][int(m)][int(l1)][int(l2)]
    
    new_data = [sum(x) for x in zip(old, datum)]
    
    full_data_set[int(r)][int(m)][int(l1)][int(l2)] = new_data
```

```python
for at_time in os.listdir('../data/'):

    for meta in os.listdir('../data/'+at_time):
        file_name = os.path.join('../data/'+at_time,meta)
        
        parts = regex.search("^data_([0-9]*)_([0-9]*)_([0-9]*)_([0-9]*)\.txt$", meta)
        
        if not parts:
            continue
        
        r  = Integer(int(parts.group(int(1))))
        m  = Integer(int(parts.group(int(2))))
        l1 = Integer(int(parts.group(int(3))))
        l2 = Integer(int(parts.group(int(4))))
        
        line = subprocess.check_output(['tail', '-1', file_name])
        
        new_data = [int(v) for v in line.split(",")[:int(2)]]
        
        add_data(r,m,l1,l2,new_data)

#######################################################################################

for at_time in os.listdir('../mesh/'):

    for meta in os.listdir('../mesh/'+at_time):
        file_name = os.path.join('../mesh/'+at_time,meta)
        
        parts = regex.search("^data_r-([0-9]*)_m-([0-9]*)_l1-([0-9]*)_l2-([0-9]*)\.csv$", meta)
        
        if not parts:
            continue
        
        r  = Integer(int(parts.group(int(1))))
        m  = Integer(int(parts.group(int(2))))
        l1 = Integer(int(parts.group(int(3))))
        l2 = Integer(int(parts.group(int(4))))
        
        line = subprocess.check_output(['tail', '-1', file_name])
        
        new_data = [int(v) for v in line.split(",")[:int(2)]]
        
        add_data(r,m,l1,l2,new_data)
```

```python
def get_prob(r,m,l1,l2):
    data = get_data(r,m,l1,l2)
    
    if not data:
        return None
    
    return Integer(data[0])/Integer(data[1])

get_prob(20,20,20,20)
```

## Output Summary

Write the data to the output files

#### Support functions

```python
def print_header(f):
    column_names = [
        "$r$", "$\ell_1$", "$\ell_2$", "$m$", "tests", "experimental", "lower", "upper"
    ]

    f.write("\\begin{tabular}{rrrr\n")
    f.write("S[table-parse-only,table-number-alignment = right]\n")
    f.write("S[table-parse-only,table-number-alignment = left]\n")
    f.write("S[table-parse-only,table-number-alignment = left]\n")
    f.write("S[table-parse-only,table-number-alignment = left]\n")
    f.write("}\n")
    f.write(" \multicolumn{1}{c}{"+("} &\n \multicolumn{1}{c}{".join(column_names))+"}\\\\\\hline\n")

def print_footer(f):
    f.write("\\hline\n\\end{tabular}")
    f.close()

def convertScientific(val):
    if regex.search('^0.0*$', val):
        return "0"
    
    parts = regex.search('^(0[.]0*)([1-9])([0-9]*)$', val)
    if parts:
        if len(parts.group(int(1))) > len("0.0"):
            if len(parts.group(int(3))) > 0:
                return parts.group(int(2))+"."+parts.group(int(3))+"e-"+str(len(parts.group(int(1)))-1)
            else:
                return parts.group(int(2))+"e-"+str(len(parts.group(int(1)))-1)
    return val
```

#### Write Output

```python
import re as regex

f = open("../output/info.txt","w")

lower0 = open("../output/lower0.txt","w")
upper0 = open("../output/upper0.txt","w")
exper0 = open("../output/exper0.txt","w")
ee     = open("../output/ee.txt","w")

for ff in [f,lower0,upper0,exper0,ee]:
    print_header(ff)

for key in data_set.keys():

    key_parts = regex.search('data_([0-9]*)_([0-9]*)_([0-9]*)_([0-9]*).txt', key)

    r  = Integer(int(key_parts.group(int(1))))
    m  = Integer(int(key_parts.group(int(2))))
    l1 = Integer(int(key_parts.group(int(3))))
    l2 = Integer(int(key_parts.group(int(4))))

    lower_bound = max_symbolic(0,lowerBound(1/6,r,l1,l2,m)).full_simplify()
    upper_bound = 0

    if (not bool(knownZeroProp(1/6,r,l,m))) and (not bool(knownZeroProp2(1/6,r,l,m))):
        upper_bound = upperBound(1/6,r,l2,m).full_simplify()

    info = [
        str(r), str(l1), str(l2), str(m),
        str(data_set[key][1]),
        convertScientific(str((Integer(data_set[key][0])/Integer(data_set[key][1])).n(digits=8))),
        convertScientific(str(lower_bound.n(digits=8))),
        convertScientific(str(upper_bound.n(digits=8)))
    ]

    is_ee = True

    if bool(lower_bound == 0):
        is_ee = False
        lower0.write(" & ".join(info)+"\\\\\n")

    if bool(upper_bound == 0):
        is_ee = False
        upper0.write(" & ".join(info)+"\\\\\n")

    if int(data_set[key][0]) == int(0):
        is_ee = False
        exper0.write(" & ".join(info)+"\\\\\n")

    if is_ee:
        ee.write(" & ".join(info)+"\\\\\n")

    f.write(" & ".join(info)+"\\\\\n")

for ff in [f,lower0,upper0,exper0,ee]:
    print_footer(ff)
    ff.close()
```

## Colour Schemes Used in Plots

The following colour scheme is based on a technical note from SRON (Paul Tol, 2012)

```python
import numpy as np
import matplotlib

PaulTol_rainbow = matplotlib.colors.ListedColormap([ \
    [ \
        np.clip( ( 0.472 - 0.567*x + 4.05*x**2 ) / ( 1 + 8.72*x - 19.17*x**2 + 14.1*x**3 ) ,0.0,1.0), \
        np.clip( 0.108932 - 1.22635*x + 27.284*x**2 - 98.577*x**3 + 163.3*x**4 - 131.395*x**5 + 40.634*x**6 ,0.0,1.0), \
        np.clip( 1 / ( 1.97 + 3.54*x - 68.5*x**2 + 243*x**3 - 297*x**4 + 125*x**5 ) ,0.0,1.0) \
    ] \
    for x in [ float(n)/float(2560-1) for n in range(int(2560)) ] \
])
```

## Contour Plots of Theoretic Bounds

### Part 1: r vs. m

#### Generate the data to produce the contour plot

```python
import matplotlib
import numpy as np
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt

length = 20

X_rm = range(6,20+1) # range of r to use
Y_rm = range(2,100+1) # range of m to use

Z1_rm = []
Z2_rm = []

for r in X_rm:
    for m in Y_rm:
        lower_bound = lowerBoundCached2(1/6,r,length,length,m,8)
        
        upper_bound = upperBoundCached2(1/6,r,length,m,8)
        
        Z1_rm = Z1_rm + [lower_bound]
        Z2_rm = Z2_rm + [upper_bound]
        

Z1_rm = np.array(Z1_rm).reshape((len(X_rm),len(Y_rm))).transpose()
Z2_rm = np.array(Z2_rm).reshape((len(X_rm),len(Y_rm))).transpose()

X_rm = np.array(X_rm)
Y_rm = np.array(Y_rm)
```

#### Plot the data

```python

cs1 = plt.contour(X_rm,Y_rm,Z1_rm)
plt.title("Lower Bound")
plt.clabel(cs1)
plt.show()

cs2 = plt.contour(X_rm,Y_rm,Z2_rm)
plt.title("Upper Bound")
plt.clabel(cs2)
plt.show()
```

```python
plt.pcolor(X_rm,Y_rm,Z1_rm,cmap=PaulTol_rainbow,vmin=0.0, vmax=1.0)
plt.title("Lower Bound")
plt.colorbar()
plt.show()

plt.pcolor(X_rm,Y_rm,Z2_rm,cmap=PaulTol_rainbow,vmin=0.0, vmax=1.0)
plt.title("Upper Bound")
plt.colorbar()
plt.show()
```
